```python
import numpy as np
import pandas as pd
import pickle as pkl
```

```python
path = "../../../Google Drive/Data_science/NYU/Machine Learning/ML Project (Collisions)/" #Joe
# path = "../../../../Google Drive/ML Project (Collisions)/" # Joyce
# path = "" # Lucas
```

```python
df = pd.read_csv(path + "NYPD_Motor_Vehicle_Collisions.csv")
```

```python
infered = pkl.load(open(path + 'infered.pkl','rb'))
```

```python
infered.set_index(infered.index.astype(int),inplace=True)
```

```python
joined = df.join(infered, on = 'UNIQUE KEY')
```

```python
joined['LATITUDE'] = joined['LATITUDE'].fillna(joined['Infered_Lat'])
joined['LONGITUDE'] = joined['LONGITUDE'].fillna(joined['Infered_Long'])
```

```python
def infered_fudge(x):
    if not(pd.isnull(x['LOCATION'])):
        return 'Original'
    elif pd.isnull(x['Infered_Lat']):
        return 'Unable_to_infer'
    else:
        return x['Lat/Long Infered']
```

```python
joined['Lat/Long Infered'] = joined.apply(infered_fudge,axis=1)
```

```python
def location_fudge(x):
    if not(pd.isnull(x['LOCATION'])):
        return x['LOCATION']
    elif not(pd.isnull(x['LATITUDE'])):
        return '({0}, {1})'.format(x['LATITUDE'],x['LONGITUDE'])
    else:
        return np.nan
```

```python
joined['LOCATION'] = joined.apply(location_fudge,axis=1)
```

```python
joined.drop(['Infered_Lat','Infered_Long'],axis=1, inplace=True)
```

```python
joined.to_csv(path+"NYPD_Motor_Vehicle_Collisions_LATLONGCLEANED.csv")
```
